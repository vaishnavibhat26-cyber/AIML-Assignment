```python
## Semantic Meaning
import warnings
warnings.filterwarnings('ignore')

word_pairs = [
    ("happy", "joyful"),
    ("big", "large"),
    ("car", "vehicle"),
    ("teacher", "student"),
    ("hot", "cold")
]

explanations = {
    ("happy", "joyful"): "These words are very similar because both express positive emotions.",
    ("big", "large"): "These words are similar because both describe size.",
    ("car", "vehicle"): "A car is a type of vehicle, so they are related in meaning.",
    ("teacher", "student"): "They are related through education, but not similar in meaning.",
    ("hot", "cold"): "These words are opposite in meaning."
}

for pair in word_pairs:
    print(f"Word Pair: {pair[0]} - {pair[1]}")
    print("Explanation:", explanations[pair])
    print("-" * 50)
```
