```python
## NumPy Speed Test
import time
import numpy as np

n = 1000000

start = time.time()
py_list = list(range(n))
py_result = [x * 2 for x in py_list]
end = time.time()
print("Python List Time:", end - start)

start = time.time()
np_array = np.arange(n)
np_result = np_array * 2
end = time.time()
print("NumPy Array Time:", end - start)
```

```python
3 Obseervations:
1.NumPy arrays are faster than Python lists for large data.
2.NumPy performs operations using optimized C-level code.
3.NumPy uses vectorized operations, reducing loop overhead.
```
