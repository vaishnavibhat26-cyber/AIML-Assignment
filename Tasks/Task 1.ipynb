```python
## 1.Print the number is odd or even 
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
## Print numbers from 10 to 1
for i in range(10, 0, -1):
    print(i)
```

```python
## Sum until user enters 0
total = 0

while True:
    num = int(input("Enter number (0 to stop): "))
    
    if num == 0:
        break
    
    total += num

print("Sum is:", total)
```
