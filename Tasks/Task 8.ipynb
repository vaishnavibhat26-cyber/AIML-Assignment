```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
data = load_breast_cancer()
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
euclidean_acc = []
k_values = range(1, 16)
```

```python
for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    euclidean_acc.append(accuracy_score(y_test, y_pred))
```

```python
manhattan_acc = []
for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k, metric='manhattan')
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    manhattan_acc.append(accuracy_score(y_test, y_pred))
```

```python
print("K   Euclidean   Manhattan")
for i in range(len(k_values)):
    print(k_values[i], "   ", euclidean_acc[i], "   ", manhattan_acc[i])
```

```python
plt.plot(k_values, euclidean_acc, marker='o', label='Euclidean')
plt.plot(k_values, manhattan_acc, marker='s', label='Manhattan')
plt.xlabel("K value")
plt.ylabel("Accuracy")
plt.title("Accuracy vs K")
plt.legend()
plt.show()
```

```python
## Explain why accuracy changes with K
Because small K is sensitive to noise and large K makes the model too general. The best K is the one that gives maximum test accuracy.
```
