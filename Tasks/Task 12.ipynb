```python
import random

text = """machine learning is very interesting and useful
learning models can generate text based on data
text generation is part of natural language processing
machine learning helps in prediction and automation"""

words = text.split()

# build word pairs (simple model)
word_dict = {}

for i in range(len(words) - 1):
    key = words[i]
    next_word = words[i + 1]
    
    if key not in word_dict:
        word_dict[key] = []
    
    word_dict[key].append(next_word)

start_word = input("Enter a starting word: ").lower()

if start_word not in word_dict:
    start_word = random.choice(list(word_dict.keys()))

sentence = [start_word]

for _ in range(9):
    current_word = sentence[-1]
    
    if current_word in word_dict:
        next_word = random.choice(word_dict[current_word])
    else:
        next_word = random.choice(list(word_dict.keys()))
    
    sentence.append(next_word)

print("Generated Sentence:")
print(" ".join(sentence))
```
