```python
import pandas as pd

data = {
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "revenue": [20000, 25000, 30000, 28000, 35000, 40000],
    "marketing": [5000, 6000, 7000, 6500, 8000, 9000],
    "profit": [15000, 19000, 23000, 21500, 27000, 31000]
}

df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt

plt.plot(df["month"], df["revenue"])
plt.title("Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.scatter(df["marketing"], df["profit"])
plt.xlabel("Marketing Spend")
plt.ylabel("Profit")
plt.title("Marketing vs Profit")
plt.show()
```

```python
print(df.corr(numeric_only=True))
```

```python
## insights of graphs 
##1.Revenue is increasing over months
##2.Higher marketing spend leads to higher profit
##3.Strong positive correlation between revenue, marketing, and profit
```
