```python
## NLP Mini App
import warnings
warnings.filterwarnings('ignore')
import re
from collections import Counter
```

```python
text = """
Natural Language Processing is a field of Artificial Intelligence.
It helps computers understand, analyze, and generate human language.
NLP is used in chatbots, translation, sentiment analysis, and speech recognition.
"""
```

```python
text = text.lower()
```

```python
text = re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
words = text.split()
```

```python
stopwords = {
    "is", "a", "of", "it", "in", "and", "the", "to", "an", "for", "on",
    "at", "by", "with", "this", "that", "from", "as"
}

filtered_words = [word for word in words if word not in stopwords]
```

```python
word_count = Counter(filtered_words)
```

```python
print("Top Keywords:")
for word, count in word_count.most_common(5):
    print(word, ":", count)
```
