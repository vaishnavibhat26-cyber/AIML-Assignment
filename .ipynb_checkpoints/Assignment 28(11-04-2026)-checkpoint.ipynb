```python
## Detection Brainstorm
5 Uses of Face/Object Detection : 
1.Face Unlock in Mobile Phones : Used to identify the user’s face and unlock the device securely.
2.Attendance System : Detects and recognizes students or employees automatically.
3.Traffic Monitoring : Detects vehicles, number plates, and pedestrians on roads.
4.Security Surveillance : Detects suspicious people or objects in public places.
5.Retail Stores : Detects customers and products for smart billing and behavior analysis.
```

```python
##One Solution Design : Face Detection Based Attendance System 
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("face.jpg")

if image is None:
    print("Image not found ❌")
else:
   
    face_cascade = cv2.CascadeClassifier(
        cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
    )

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    for (x, y, w, h) in faces:
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)

    print("Number of faces detected:", len(faces))

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb)
    plt.title("Face Detection Output")
    plt.axis("off")
    plt.show()
```
