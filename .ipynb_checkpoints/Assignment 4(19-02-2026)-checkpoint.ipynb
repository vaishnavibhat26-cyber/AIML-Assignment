```python
## Student Data Manager
students = {
    "A": 85,
    "B": 72,
    "C": 90,
    "D": 65,
    "E": 78
}

topper = max(students, key=students.get)
print("Topper:", topper, "-", students[topper])

average = sum(students.values()) / len(students)
print("Class Average:", average)

print("\nGrades:")
for name, marks in students.items():
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 60:
        grade = "C"
    else:
        grade = "D"
    
    print(name, ":", marks, "-> Grade", grade)
```

```python
## Student Data Manager
students = {
    "Student1": {"name": "Ravi", "marks": 85},
    "Student2": {"name": "Anu", "marks": 25},
    "Student3": {"name": "Kiran", "marks": 78},
    "Student4": {"name": "Meena", "marks": 88},
    "Student5": {"name": "Arjun", "marks": 95}
}

topper = max(students.values(), key=lambda x: x["marks"])
print("Topper:", topper["name"], "-", topper["marks"])
total = sum(student["marks"] for student in students.values())
average = total / len(students)
print("Class Average:", average)
print("\nGrades:")
for student in students.values():
    marks = student["marks"]
    
    if marks >= 90:
        grade = "A"
    elif marks >= 75:
        grade = "B"
    elif marks >= 50:
        grade = "C"
    else:
        grade = "Fail"
        
    print(student["name"], ":", grade)
```
