```python
## Break the AI
import warnings
warnings.filterwarnings('ignore')

prompts = [
    "What is heavier, 1 kg of iron or 1 kg of cotton?",
    "If five machines make five items in five minutes, how long do 100 machines take to make 100 items?",
    "Can a man marry his widow’s sister?",
    "Repeat the word poem 20 times and then answer: what do cows drink?",
    "A farmer has 17 sheep and all but 9 die. How many are left?"
]
```

```python
results = [
    "Both are equal in weight.",
    "5 minutes.",
    "No, because he is dead.",
    "Cows drink water.",
    "9 sheep are left."
]
```

```python
for i in range(len(prompts)):
    print(f"Prompt {i+1}: {prompts[i]}")
    print(f"Result: {results[i]}")
    print("-" * 60)
```
