```python
##Movie Review Analyzer
import warnings
warnings.filterwarnings('ignore')

reviews = [
    "The movie was amazing and the acting was excellent",
    "The film was boring and a complete waste of time",
    "I love the music and the story was great",
    "The plot was dull and the ending was terrible",
    "The movie was good but a little long"
]
```

```python
positive_words = ["good", "excellent", "amazing", "great", "enjoyable", "fantastic", "love"]
negative_words = ["bad", "boring", "poor", "terrible", "waste", "dull", "disappointing"]
```

```python
def sentiment_analyzer(review):
    review = review.lower()
    pos_count = sum(word in review for word in positive_words)
    neg_count = sum(word in review for word in negative_words)
    
    if pos_count > neg_count:
        return "Positive"
    elif neg_count > pos_count:
        return "Negative"
    else:
        return "Neutral"
```

```python
for i, review in enumerate(reviews, 1):
    result = sentiment_analyzer(review)
    print(f"Review {i}: {review}")
    print("Sentiment:", result)
    print("-" * 50)
```
