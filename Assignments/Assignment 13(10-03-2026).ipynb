```python
## Spam Classifier Thinking
spam_system = {
    "Features": [
        "message text",
        "number of links",
        "keywords (free, win)",
        "message length",
        "sender info"
    ],
    
    "Data Needed": [
        "labeled messages (spam / not spam)",
        "large dataset of emails/SMS"
    ],
    
    "Mistakes": [
        "false positive (normal → spam)",
        "false negative (spam → normal)"
    ]
}
for key, value in spam_system.items():
    print("\n", key)
    for v in value:
        print("-", v)
```

```python
##Spam Classifier Thinking
1. Features (Input)
Message text
Number of links
Presence of keywords (free, win, offer)
Message length
Sender information

2. Data Needed
Large dataset of SMS/emails
Labels → Spam / Not Spam
Example:
“Win a free iPhone” → Spam
“Meeting at 5 pm” → Not Spam

3. Possible Mistakes
False Positive → Normal message marked as spam 
False Negative → Spam message not detected 

4. Simple Explanation
ML learns patterns from messages
If message looks similar to spam → classified as spam
```
