```python
## Dataset Detective
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Marks": [85, 90, 78, 92, 88],
    "Age": [20, 21, 19, 22, 20]
}

df = pd.DataFrame(data)

print("Top rows:")
print(df.head())

print("\nHighest values:")
print(df.max(numeric_only=True))

print("\nMissing values:")
print(df.isnull().sum())
```
