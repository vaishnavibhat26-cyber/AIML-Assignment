```python
weak_prompts = {
    "Resume": "Make a resume.",
    "Business Idea": "Give me a business idea.",
    "Study Plan": "Make a study plan."
}

strong_prompts = {
    "Resume": "Create a professional one-page resume for a final-year Computer Science student skilled in Python, HTML, CSS, JavaScript, and machine learning. Include education, technical skills, projects, internships, and certifications.",
    
    "Business Idea": "Suggest 3 low-investment business ideas for a college student in India interested in online services, digital marketing, and AI tools. Include target customers, budget, and earning potential.",
    
    "Study Plan": "Create a 7-day study plan for a Computer Science student preparing for exams in Python, DBMS, and Computer Networks. Include 3 hours of study per day, revision time, and breaks."
}
```

```python
for category in weak_prompts:
    print(f"\nCategory: {category}")
    print("Weak Prompt :", weak_prompts[category])
    print("Strong Prompt:", strong_prompts[category])
    print("-" * 70)
```
