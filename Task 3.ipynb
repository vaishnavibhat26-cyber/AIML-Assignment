```python
import pandas as pd

data = {
    "name": ["A", "B", "C", "D", "E"],
    "maths": [80, 60, 90, 50, 78],
    "science": [75, 70, 88, 0, 82],
    "english": [90, 65, 92, 60, 85],
    "dept": ["CSE", "ISE", "CSE", "ECE", "ISE"]
}

df = pd.DataFrame(data)
df.to_csv("students.csv", index=False)

print("students.csv created successfully")
```

```python
import pandas as pd

df = pd.read_csv("students.csv")
df = df.fillna(0)
df["total"] = df["maths"] + df["science"] + df["english"]
df["average"] = df["total"] / 3

top3 = df.sort_values(by="total", ascending=False).head(3)
print("Top 3 Students:\n", top3)

dept_avg = df.groupby("dept")[["maths", "science", "english"]].mean()
print("\n Department-wise Average:\n", dept_avg)

above_75 = df[
    (df["maths"] > 75) &
    (df["science"] > 75) &
    (df["english"] > 75)
]
print("\nStudents scoring above 75 in all subjects:\n", above_75)

print("\nUpdated Data:\n", df)
```
